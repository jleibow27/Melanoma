```python
#path to the base dir
base_dir = "../data/DermMel/train"

#path of various folders
train_Melanoma = os.path.join(base_dir, "../data/DermMel/train/Melanoma")
train_NotMelanoma = os.path.join(base_dir, "../data/DermMel/train/NotMelanoma")

test_Melanoma = os.path.join(base_dir, "../data/DermMel/test/Melanoma")
test_NotMelanoma = os.path.join(base_dir, "../data/DermMel/test/NotMelanoma")

Not_Melanoma = train_NotMelanoma + test_NotMelanoma
Melanoma = train_Melanoma + test_Melanoma


#list of available labels
classes = ["NotMelanoma", "Melanoma"]

dataset = []
for category in classes:
    path = os.path.join(base_dir, category)
    for img in os.listdir(path):
        #`enter code here`print(img)
        image = cv2.imread(os.path.join(path,img))
        label = classes.index(category)
        dataset.append([image,label])
print(dataset[1])
```
